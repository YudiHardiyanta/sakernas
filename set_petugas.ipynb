```python
import pandas as pd
import os
```

```python
files = os.listdir('input_ipds')
```

```python
import json
```

```python
dt_all = []
for f in files:
    with open(f'input_ipds/{f}', 'r') as file:
        data = json.load(file)
        petugas = pd.json_normalize(data['searchData'])
        dt_all.append(petugas)
dt_all = pd.concat(dt_all, ignore_index=True)
```

```python
dt_all=dt_all[['id','codeIdentity',
               'region.level1.level2.name',
               'region.level1.level2.level3.name',
               'region.level1.level2.level3.level4.name',
               'region.level1.level2.level3.level4.level5.name',
               'data1','data2',
 'data3',
 'data4',
 'data5','currentUserUsername',
 'currentUserFullname',
 'currentUserSurveyRoleName','longitude','latitude']].rename(columns={
    'region.level1.level2.name' : 'kab_name',
               'region.level1.level2.level3.name' : 'kec_name',
               'region.level1.level2.level3.level4.name' : 'desa_name',
               'region.level1.level2.level3.level4.level5.name' : 'sls_name',
    'data1':'NUS','data2':'KRT','data3':'NO_BANG','data4' : 'alamat','currentUserUsername' : 'email','currentUserFullname': 'petugas','currentUserSurveyRoleName' : 'role'})
```

```python
dt_all['KODE_WILAYAH']=dt_all['codeIdentity'].str[:16]
```

```python
dt_all.to_excel('petugas.xlsx', index=False)
```
